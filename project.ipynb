#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

```python
# TODO
from scipy.stats import zscore

#Function to normalize
#normalize = lambda x: (x - x.min()) / (x.max() - x.min())

south_belt_madrid_subset.loc[:, 'normalized_price'] = south_belt_madrid_subset.groupby('level5')['price'].transform(zscore)

#Plot histograms in same plot
plt.figure(figsize=(10,6))
for place in south_belt_madrid:
    mini_set = south_belt_madrid_subset[south_belt_madrid_subset['level5'] == place]
    plt.hist(mini_set['normalized_price'], bins=15, alpha=0.5, label=place, edgecolor= "white")
    plt.title('Normalized price in South Belt of Madrid')
    plt.xlabel('Normalized price')
    plt.ylabel('Frquency')
    plt.legend()
    plt.show
```

```python
# TODO

#Function to normalize
#normalize = lambda x: (x - x.min()) / (x.max() - x.min())
normalize = lambda x: (x - x.mean()) / (x.std())

south_belt_madrid_subset.loc[:, 'normalized_price'] = south_belt_madrid_subset.groupby('level5')['price'].transform(normalize)

#Plot histograms in same plot
plt.figure(figsize=(10,6))
for place in south_belt_madrid:
    mini_set = south_belt_madrid_subset[south_belt_madrid_subset['level5'] == place]
    plt.hist(mini_set['normalized_price'], bins=15, alpha=0.5, label=place, edgecolor= "white")
    plt.title('Normalized price in South Belt of Madrid')
    plt.xlabel('Normalized price')
    plt.ylabel('Frquency')
    plt.legend()
    plt.show
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

By looking at the price per square meter (PPS) for houses in Getafe and Alcorcón, we can compare how expensive these towns are.

Alcorcón's average PPS is higher than Getafe's. This means that houses in Alcorcón are more expensive pps.

Regarding variability we can see that prices differ and Getafe has more variability, which measn that Getafe has a wider range of prices.

```python
from ipyleaflet import Map, basemaps, CircleMarker

# Map centered on (60 degrees latitude and -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (40.4168, -3.7038), zoom = 10, min_zoom = 1, max_zoom = 20, basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
south_belt_madrid_subset.loc[:, 'latitude'] = south_belt_madrid_subset['latitude'].apply(lambda x: float(x.replace(',', '.')))
south_belt_madrid_subset.loc[:, 'longitude'] = south_belt_madrid_subset['longitude'].apply(lambda x: float(x.replace(',', '.')))

colors = {"Fuenlabrada": 'orange', "Leganés": 'blue', "Getafe": 'green', "Alcorcón": 'purple'}

for _, row in south_belt_madrid_subset.iterrows():
    marker = CircleMarker(location=(row['latitude'], row['longitude']), radius=4, color=colors[row['level5']], fill=True, fill_color=colors[row['level5']])
    map.add_layer(marker)
```
